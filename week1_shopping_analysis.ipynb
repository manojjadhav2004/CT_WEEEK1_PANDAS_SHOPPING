## Step 1: Load CSV into a Pandas DataFrame

```python
import pandas as pd

df = pd.read_csv('Combined_dataset.csv')
print('Dataset loaded successfully!')
print(f'Shape: {df.shape}')
```

## Step 2: Explore the Data

```python
df.head()
```

```python
df.tail()
```

```python
print('Shape (rows, columns):', df.shape)
print('\nColumn names:')
print(df.columns.tolist())
```

```python
print('Data Types:')
print(df.dtypes)
```

## Step 3: Handle Missing Values

```python
print('Missing values per column:')
print(df.isnull().sum())
```

```python
# Drop columns that have too many missing values (not useful for analysis)
df.drop(columns=['videos', 'what_customers_said', 'seller_name',
                  'seller_information', 'variations'], inplace=True)

# Fill missing 'discount' with 0 (no discount)
df['discount'].fillna(0, inplace=True)

print('Missing values after handling:')
print(df.isnull().sum())
```

## Step 4: Basic Operations — Filter Rows & Select Columns

```python
# Select specific columns
selected = df[['product_id', 'title', 'category', 'rating', 'initial_price', 'discount']]
print('Selected columns:')
selected.head()
```

```python
# Filter rows: products with rating >= 4.0
high_rated = df[df['rating'] >= 4.0]
print(f'Products with rating >= 4.0: {len(high_rated)}')
high_rated[['title', 'category', 'rating', 'initial_price']].head()
```

## Step 5: Remove Duplicates

```python
print(f'Rows before removing duplicates: {len(df)}')
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
print(f'Rows after removing duplicates : {len(df)}')
```

## Step 6: Create Derived Column — total_amount

```python
# total_amount = initial_price * ratings_count
# (price * quantity — using ratings_count as a proxy for quantity sold)
df['total_amount'] = df['initial_price'] * df['ratings_count']

print('total_amount column created:')
df[['title', 'initial_price', 'ratings_count', 'total_amount']].head()
```

## Step 7: Save Cleaned Dataset as New CSV

```python
df.to_csv('shopping_cleaned.csv', index=False)
print('Cleaned dataset saved as shopping_cleaned.csv')
print(f'Final shape: {df.shape}')
```
